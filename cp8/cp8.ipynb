#### e) Tomando $V_D=0.7$, cuando $V_i>4.7$, Diodo ON, $V_o=4.7$, en el resto $V_o=V_i$

```python
f=60
t=np.linspace(0,100e-3,1001)

v_i=8*np.cos(2*np.pi*f*t)
v_o=[vi if vi<4.7 else 4.7 for vi in v_i]
plt.plot(t,v_i)
plt.plot(t,v_o,color='red')
plt.xlabel('t(s)')
plt.ylabel('V')
plt.legend(['vi','vo'],loc='upper right')
plt.grid(True)
```

#### f) Tomando $V_D=0.7$, cuando $V_i>3.7$, Diodo ON, $V_o=0.7$, en el resto $V_o=V_i-3$

```python
f=60
t=np.linspace(0,100e-3,1001)

v_i=8*np.cos(2*np.pi*f*t)
v_o=[vi-3 if vi<3.7 else 0.7 for vi in v_i]
plt.plot(t,v_i)
plt.plot(t,v_o,color='red')
plt.xlabel('t(s)')
plt.ylabel('V')
plt.legend(['vi','vo'],loc='upper right')
plt.grid(True)
```
